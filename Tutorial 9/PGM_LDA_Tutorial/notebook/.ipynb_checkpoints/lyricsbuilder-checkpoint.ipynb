```python
N = 100
workbook = xlrd.open_workbook("/home/manttari/Documents/PGM/tutorial/dataset/MoodyLyrics/ml_balanced.xlsx")
sheet = workbook.sheet_by_index(0)
ATE = []
for rowx in range(16,sheet.nrows):
    cols = np.array(sheet.row_values(rowx))
    ATE.append(cols[[1,2,3]])
    #print (ATE[0])
```

```python
testDocs = {'groundTruth':[], 'lyrics':[]}
for i in range(int(N*0.8),N):
    fetchedLyrics = ""
    try:
        fetchedLyrics = lyrics.getlyrics(ATE[i][0],ATE[i][1])
        if(fetchedLyrics!=False):
            testDocs['groundTruth'].append(ATE[i][2])
            testDocs['lyrics'].append(fetchedLyrics)
    except IOError: 
        print("couldn't find lyrics for: ",ATE[i][0],"-",ATE[i][1])
        
    if((i+1)%100==0):
        print(i+1, " out of ",N)
   
pickle.dump( testDocs, open( "testLyricsSuperSmall2.p", "wb" ) )
#vectorizer = pickle.load(open( "vectorizerSuperSmall.p", "rb" ) )

#y = vectorizer.transform(testDocs['lyrics'])
#y
```

```python
apDocs = []
inContent = False
currentDoc = ""
docs = open("/home/manttari/Documents/PGM/tutorial/dataset/ap/ap.txt", "r") 
for line in docs.readlines(): 
    if("</TEXT>" in line):
        inContent=False
        apDocs.append(currentDoc)
    if(inContent):
        currentDoc += line
    if("<TEXT>" in line):
        inContent=True
        currentDoc = ""
    
derp = len(apDocs)
trainApDocs = apDocs[:derp/10]
testApDocs = apDocs[derp/10: derp/10 + 50]
allApDocs = trainApDocs + testApDocs
print(derp)
pickle.dump(trainApDocs, open("trainApDocsSuperSmall2.p", "wb"))
pickle.dump(testApDocs, open("testApDocsSuperSmall2.p", "wb"))
print(len(trainApDocs))
print(len(testApDocs))
print(len(allApDocs))
```

```python
vectorizer = CountVectorizer(max_df=0.5, min_df=0.01, stop_words='english', max_features=750)
x = vectorizer.fit(allApDocs)
print(len(x.get_feature_names()))
pickle.dump( vectorizer, open( "vectorizerAPSuperSmall2.p", "wb" ) )
```
